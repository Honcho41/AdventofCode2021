```python
with open("depth.txt") as f:
    depth = [int(x) for x in f.read().split()]
```

```python
def sliding_window(elements, window_size):
    if len(elements) <= window_size:
        return elements
    for i in range(len(elements) - window_size + 1):
        print(elements[i:i+window_size])
        
sliding_window(depth, 3)
```

```python
increase = []

def increase_calc(elements, window_size):
    if len(elements) <= window_size:
        return elements
    for i in range(len(elements) - window_size + 1):
        increase.append(elements[i:i+window_size])
        print(elements[i:i+window_size])
        
```

```python
increase_calc(depth, 3)
```
